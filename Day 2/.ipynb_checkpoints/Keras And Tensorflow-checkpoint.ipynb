# Keras And Tensorflow

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.linspace(1, 100, 10000)/1000
y = 2 * x
```

```python
model1 = Sequential()
model1.add(Dense(1, input_dim = 1, activation = 'linear'))
model1.add(Dense(1)) #output layer
model1.summary()
```

```python
model1.compile('sgd', 'mse', metrics=['mae', 'acc'])
```

```python
history = model1.fit(x, y, nb_epoch = 1000, verbose = 1)
```

```python
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['mae'])
plt.plot(history.history['acc'])
plt.title('model loss')
plt.legend(['loss', 'mae', 'acc'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
